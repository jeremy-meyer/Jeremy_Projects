Basic Plotting
R's equilavent of lines: Just use plt.plot() twice before plt.show()

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4]) #x index is 0,1,2...
plt.plot(range(4), [1, 4, 9, 16], color=(2/5,2/5,1)) #Use this if you want to specify points
                                                     #Can use a RGB tuple for numbered colors
plt.ylabel('some numbers')
plt.xlabel("Index")
plt.show()
```

Plotting with different colors and styles
    List of colors:
    https://matplotlib.org/examples/color/named_colors.html


```python
import numpy as np
def f(x):
    return(np.sin(x)) #Explicit return statement

xs = np.arange(0,2*np.pi, 0.1) #Arange for decimal increments

plt.plot(xs, [f(x) for x in xs], color='greenyellow', linestyle=':') 
plt.plot(xs, [f(x-2) for x in xs], 'c--', linewidth=2) #Cyan Dashed line. Witdh of 2
plt.show()
```

PLotting markers and specifying plot limits
    use marker='o' for points. More markers:
    https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4], [1,4,9,16], 'ro') #Red Circles
plt.plot([1,2,3,4], [16,9,4,1], marker='.', linestyle='none') #Removes connecting line
plt.axis([0, 6, 0, 20]) #xlim(0,6), ylim(0,20)
#plt.xlim(0,6)
#plt.ylim(0,20)
plt.show()
```

Plot multiple lines in one plot statement. plt.plot() can take multiple arguements

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
